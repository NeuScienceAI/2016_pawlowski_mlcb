```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
floxuridine = pd.read_csv('../../input/floxuridine.csv')
plt.plot(floxuridine['concentration'], floxuridine['distance'], 'o--')
plt.xlabel('Concentration')
plt.ylabel('Distance from DMSO')
plt.title('Dose response of Floxuridine')
plt.gca().set_xscale('log')
plt.show()
```
